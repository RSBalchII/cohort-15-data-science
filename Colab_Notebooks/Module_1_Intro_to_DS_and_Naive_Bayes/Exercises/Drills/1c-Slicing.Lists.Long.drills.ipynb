# List for Slicing Generator

The code below generates a list of numbers with the words "dog", "cat", and "bat" embedded, in that order.  Your task is three fold:
1. pick a number for the seed between 0 and 49, inclusive
1. run the gen_my_list() using the seed
1. display the slice syntax ( i.e. start, stop, skip values ) for slicing out the words "dog", "cat", and "bat", in that order.

```python
import random
def gen_my_list(seed):
    '''Provide an integer seed to randomly create a list of numbers'''
    random.seed(seed)
    prefix = list(range(4, random.randint(100,2000) ))
    random.shuffle(prefix)
    suffix = list(range(4, random.randint(100,2000) ))
    random.shuffle(suffix)
    tens = [10] * ( random.randint(100,500) * 2 + 1)
    tens[0::len(tens)//2] = ["dog","cat","bat"]
    tens = tens[::(random.randint(0,1)*2-1)]
    return prefix + tens + suffix
```

## Example with seed 0


```python
seed = 0
my_list = gen_my_list(seed)
", ".join([ str(x) for x in my_list ])
```
