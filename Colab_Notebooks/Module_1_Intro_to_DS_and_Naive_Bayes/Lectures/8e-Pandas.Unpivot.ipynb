```python
import pandas as pd
import markdown
```

```python
# Assign Markdown string to a variable
table_md = '''|Date|A|B|C|D|E |
|----|-|-|-|-|--|
|2005|1|2|3|4|50|
|2006|6|7|8|9|10|'''
table_md
```

```python
# Convert Markdown to HTML to a DataFrame
table_html=markdown.markdown(table_md, extensions=['markdown.extensions.tables'])
table_html
```

```python
%%html
<table>
<thead>
<tr>
<th>Date</th>
<th>A</th>
<th>B</th>
<th>C</th>
<th>D</th>
<th>E</th>
</tr>
</thead>
<tbody>
<tr>
<td>2005</td>
<td>1</td>
<td>2</td>
<td>3</td>
<td>4</td>
<td>50</td>
</tr>
<tr>
<td>2006</td>
<td>6</td>
<td>7</td>
<td>8</td>
<td>9</td>
<td>10</td>
</tr>
</tbody>
</table>
```

```python
df = pd.read_html(table_html)[0]
df
```

# Using .stack()

```python
df.set_index('Date')
```

```python
df.set_index('Date').stack()
```

```python
df.set_index('Date').stack().reset_index(name='Val')
```

```python
df_stack = df.set_index('Date').stack().reset_index(name='Val').rename(columns={'level_1':'X'})
df_stack
```

# Using .melt()

## Sorted by "X", then "Date"

```python
keeps = df.columns[:1]
keeps
```

```python
df_melt = df.melt(keeps, var_name='X', value_name='Val')
df_melt
```

## Sorted by "Date", then "X"

```python
df_melt2 = ( df
            .melt(keeps, var_name='X', value_name='Val')
            .sort_values(['Date','X'])
            .reset_index(drop=True)
)
df_melt2
```
