````python
def acronym_creator(text):
    """
    Create an acronym from the input text.
    
    Args:
        text (str): The input text to create an acronym from.
    
    Returns:
        str: The acronym created from the input text.
    """
    words = text.split()
    acronym = ''.join(word[0].upper() for word in words)
    return acronym

# Example usage
text = "My name is Nevin"
result = acronym_creator(text)
print(result)  # Output: MNIN
````

````python
def email_slicer(email):
    try:
        username, domain = email.split('@')
        return username, domain
    except ValueError:
        return None, None

# Test the functions
email = 'nevin@cnm.edu'

email_result = email_slicer(email)

print(f"Email Slicer Result: {email_result}")
````
